```python
import pandas as pd
import keras
import tensorflow as tf
from keras import backend as K
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Input, Embedding, BatchNormalization, concatenate
from keras.optimizers import Adam
from keras.layers.merge import dot
from keras.models import Model
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rating_data = pd.read_csv('final_data.csv')
```

```python
rating_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(rating_data[['user_id','battery_id']], rating_data[['rating']], test_size=0.2, random_state=42)
```

```python
latent_factors = 20
```

```python
user_input = Input(shape=(1,),name='user_input', dtype='int32')
user_embedding = Embedding(151, latent_factors, name='user_embedding')(user_input)
user_flat = Flatten(name='user_flat')(user_embedding)
user_flat = Dropout(0.2)(user_flat)
```

```python
movie_input = Input(shape=(1,),name='movie_input', dtype='int32')
movie_embedding = Embedding(40, latent_factors, name='movie_embedding')(movie_input)
movie_flat = Flatten(name='movie_flat')(movie_embedding)
movie_flat = Dropout(0.2)(movie_flat)
```

```python
product = dot([user_flat, movie_flat], name='product', axes=1)
```

```python
dense1 = Dense(512, activation='relu')(product)
dense1 = Dropout(0.2)(dense1)
dense2 = Dense(256, activation='relu')(dense1)
dense2 = Dropout(0.2)(dense2)
dense3 = Dense(128, activation='relu')(dense2)
dense3 = Dropout(0.2)(dense3)
dense4 = Dense(64, activation='relu')(dense3)
dense_out = Dense(1, activation='relu')(dense4)
```

```python
model = Model([user_input, movie_input], dense_out)
model.summary()
```

```python
learning_rate = 1e-4
epochs = 120
batch_size = 256
```

```python
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        min_delta=1e-3,
        patience=4,
        verbose=1)
]
```

```python
model.compile(optimizer=Adam(lr=learning_rate), loss='mse', metrics=[tf.keras.metrics.RootMeanSquaredError(), tf.keras.metrics.MeanAbsoluteError()])
```

```python
fit = model.fit([x_train['user_id'], x_train['battery_id']], y_train,
                batch_size=batch_size,
                epochs=epochs,
                callbacks=callbacks,
                shuffle=True, 
                validation_split=0.25)
```

```python
x_train, x_test, y_train, y_test = train_test_split(rating_data[['user_id','battery_id','side_info']], rating_data[['rating']], test_size=0.2, random_state=42)
```

```python
num_users = 150
num_batteries = 39
```

```python
aux_input = Input(shape=(1,))
aux_dense1 = Dense(128, activation='relu')(aux_input)
aux_dense1 = Dropout(0.2)(aux_dense1)
aux_dense2 = Dense(64, activation='relu')(aux_dense1)
aux_dense2.shape
```

```python
latent_factors = 480
```

```python
user_input = Input(shape=(1,),name='user_input', dtype='int32')
user_embedding = Embedding(num_users+1, latent_factors, name='user_embedding')(user_input)
user_flat = Flatten(name='user_flat')(user_embedding)
user_flat = Dropout(0.2)(user_flat)
```

```python
battery_input = Input(shape=(1,),name='battery_input', dtype='int32')
battery_embedding = Embedding(num_batteries+1, latent_factors, name='battery_embedding')(battery_input)
battery_flat = Flatten(name='battery_flat')(battery_embedding)
battery_flat = Dropout(0.2)(battery_flat)
```

```python
product = dot([user_flat, battery_flat], name='product', axes=1)
```

```python
dense1 = Dense(512, activation='relu')(product)
dense1 = Dropout(0.2)(dense1)
dense2 = Dense(256, activation='relu')(dense1)
dense2 = Dropout(0.2)(dense2)
dense3 = Dense(128, activation='relu')(dense2)
dense3 = Dropout(0.2)(dense3)
dense4 = Dense(64, activation='relu')(dense3)
```

```python
product2 = dot([dense4, aux_dense2], name ='product2', axes=1)
```

```python
model = Model([user_input,battery_input,aux_input],product2)
```

```python
model.summary()
```

```python
learning_rate = 1e-4
epochs = 120
batch_size = 256
```

```python
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        min_delta=2e-4,
        patience=4,
        verbose=1)
]
```

```python
model.compile(optimizer=Adam(lr=learning_rate), loss='mse', metrics=[tf.keras.metrics.RootMeanSquaredError(), tf.keras.metrics.MeanAbsoluteError()])
```

```python
history = model.fit([x_train['user_id'], x_train['battery_id'], x_train['side_info']], y_train,
                batch_size=batch_size,
                epochs=epochs,
                shuffle=True, 
                validation_split=0.25)
```
